```python
texto = ''.join(texto_teste)
```

```python
print(texto)
```

```python
def criar_vetorizacao():
    vetorizador = TfidfVectorizer(
        strip_accents='unicode', 
        stop_words=stopwords_lista, 
        lowercase=True, ngram_range=(1,3), 
        max_df=1.0, 
        min_df=0.02,
        max_features=50000,
    )
    return vetorizador
```

```python
vetorizador = criar_vetorizacao()
```

```python
variavel_teste = vetorizador.fit_transform([texto])
```

```python
print(variavel_teste)
```

```python
indices_palavras_frequentes = variavel_teste.toarray().argsort()[0][-10:][::-1]
palavras_frequentes1 = [vetorizador.get_feature_names()[i] for i in indices_palavras_frequentes]
```
